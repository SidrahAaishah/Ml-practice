```python
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.head(5)
```

```python
X = heart_disease.drop("target",axis = 1)
y = heart_disease["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

clf.get_params()
```

```python
clf.fit(X_train,y_train)
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds) #by default the clf.score returns the accuracy_score
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

best_score = 0
best_estimators = None

np.random.seed(35)

for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i, random_state=35).fit(X_train, y_train)
    
    test_score = clf.score(X_test, y_test)
    cv_score = np.mean(cross_val_score(clf, X_train, y_train, cv=5)) * 100
    
    print(f"Model accuracy on test set : {test_score}")
    print(f"Cross-validation score : {cv_score}%")
    
    if cv_score > best_score:
        best_score = cv_score
        best_estimators = i

print("\nBest n_estimators from CV:", best_estimators)
print("Best CV score:", best_score)

# Train final model
final_clf = RandomForestClassifier(n_estimators=best_estimators, random_state=35)
final_clf.fit(X_train, y_train)
print("Final test set accuracy:", final_clf.score(X_test, y_test))
```
